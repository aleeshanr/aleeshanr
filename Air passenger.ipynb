# "Air Passenger Travel Trends from 1949 to 1960: An Analytical Overview"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
dataset=pd.read_csv("DownloadS//AirPassengers.csv")
dataset
```

```python
#Convert to data
data=dataset.rename(columns ={'#passengers':'passengers'})
data
```

```python
#Getting Mean, Mode, Median, Sum of the Number of passengers travelled in a particular year
data_array=np.array(data)
data_list_year = []
for i in data_array:
  final_data = i[0].split("-")
  #print(final_data)
  pass_count = i[1]
  #print(pass_count)
  final_data = list(final_data)
  final_data.append(pass_count)
  data_list_year.append(final_data)

```

```python
#Converting the list into DataFrame
df = pd.DataFrame(data_list_year)
df = df.rename(columns = {0:"Year",1:"Month",2:"Passenger"})
print(df)
```

```python
fig, axes = plt.subplots(4,3,figsize =(25,15))
fig.supxlabel("year",fontsize=24)
fig.supylabel("passenger count",fontsize=24)

df_jan=df[df['Month']=="01"]
df_feb=df[df['Month']=="02"]
df_mar=df[df['Month']=="03"]
df_apr=df[df['Month']=="04"]
df_may=df[df['Month']=="05"]
df_jun=df[df['Month']=="06"]
df_jul=df[df['Month']=="07"]
df_aug=df[df['Month']=="08"]
df_sep=df[df['Month']=="09"]
df_oct=df[df['Month']=="10"]
df_nov=df[df['Month']=="11"]
df_dec=df[df['Month']=="12"]


#To plot
axes[0,0].plot(df_jan['Year'],df_jan['Passenger'])
axes[0,0].set_title("january")

axes[0,1].plot(df_feb['Year'],df_feb['Passenger'])
axes[0,1].set_title("february")

axes[0,2].plot(df_mar['Year'],df_mar['Passenger'])
axes[0,2].set_title("March")


axes[1,0].plot(df_apr['Year'],df_apr['Passenger'])
axes[1,0].set_title("April")


axes[1,1].plot(df_may['Year'],df_may['Passenger'])
axes[1,1].set_title("May")


axes[1,2].plot(df_jun['Year'],df_jun['Passenger'])
axes[1,2].set_title("june")


axes[2,0].plot(df_jul['Year'],df_jul['Passenger'])
axes[2,0].set_title("July")


axes[2,1].plot(df_aug['Year'],df_aug['Passenger'])
axes[2,1].set_title("August")


axes[2,2].plot(df_sep['Year'],df_sep['Passenger'])
axes[2,2].set_title("September")


axes[3,0].plot(df_oct['Year'],df_oct['Passenger'])
axes[3,0].set_title("October")


axes[3,1].plot(df_nov['Year'],df_nov['Passenger'])
axes[3,1].set_title("November")


axes[3,2].plot(df_dec['Year'],df_dec['Passenger'])
axes[3,2].set_title("December")

plt.subplots_adjust(hspace = 0.5)
plt.show()
```

```python

# Calculate average passenger count for each month over the years
monthly_avg_passengers = df.groupby('Month')['Passenger'].mean().sort_values()
print(monthly_avg_passengers)
# Identify most and least frequently traveled months
most_frequent_months = monthly_avg_passengers.tail(3).index.tolist()  # July, August, September
least_frequent_month = monthly_avg_passengers.head(1).index.tolist()[0]  # November

# Plot monthly trends over the years
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Passenger', hue='Month', ci=None)
plt.title('Monthly Passenger Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Passenger Count')

# Highlight most and least frequently traveled months
for month in most_frequent_months:
    plt.axhline(y=monthly_avg_passengers[month], color='green', linestyle='--', label='Most Frequent Months')
plt.axhline(y=monthly_avg_passengers[least_frequent_month], color='red', linestyle='--', label='Least Frequent Month')

plt.legend()
plt.show()

# Plot overall passenger trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Passenger', estimator='sum')
plt.title('Overall Passenger Trend (1949-1960)')
plt.xlabel('Year')
plt.ylabel('Total Passenger Count')
plt.show()

# Calculate and print annual growth rate
total_passengers_by_year = df.groupby('Year')['Passenger'].sum()
growth_rate = ((total_passengers_by_year.iloc[-1] / total_passengers_by_year.iloc[0]) ** (1 / len(total_passengers_by_year))) - 1
print(f"Annual Growth Rate: {growth_rate:.2%}")
```
